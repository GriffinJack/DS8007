# Image 1
#### Link : https://i1.wp.com/eagereyes.org/media/2008/nytimes-mrspresident.jpg
* If the numbers don't add up to a 100 assume the rest  were "netural," i.e., did not answer

![Image 1](https://i1.wp.com/eagereyes.org/media/2008/nytimes-mrspresident.jpg)

For this image I feel like the use of color was underutilized. They could have colored the circles to show yes as green and no as red. This could add to the story being told as we are focusing on the yes and no answers to the question. The colors could allow for a preattentive visual as it would be much more obvious as to what the image is representing. This data could also be read easier by changing each category to be a pie chart instead. This would allow for easier comparison between yes and no responses. This representation also leaves out information such as the netural results. 

```python
mrsPresident = {'Under 45':[79,16, (100-(79+16))], '45-64':[69,22, (100-(69+22))], '65+': [44,39, (100-(44+39))], 'Men':[65,25, (100-(65+25))], 'Women':[72,20, (100-(72+20))]}
labels = ['yes','no','netural']
colors=['green','red','yellow']
fig = plt.figure()

#U45
axU45 = fig.add_axes([0,0,0.5,0.5], aspect=1)
axU45.pie(mrsPresident['Under 45'], labels=labels, colors=colors, autopct='%.1f%%')
axU45.set_title("Under 45")


#45-64
ax45_65 = fig.add_axes([0,-.6,0.5,0.5], aspect=1)
ax45_65.pie(mrsPresident['45-64'], labels=labels, colors=colors, autopct='%.1f%%')
ax45_65.set_title('45-64')

#64+
ax64 = fig.add_axes([0,-1.2,0.5,0.5])
ax64.pie(mrsPresident['65+'], labels=labels, colors=colors, autopct='%.1f%%')
ax64.set_title('65+')

#Men
axMen = fig.add_axes([0,-1.8,0.5,0.5])
axMen.pie(mrsPresident['Men'], labels=labels, colors=colors, autopct='%.1f%%')
axMen.set_title('Men')

#Women
axWomen = fig.add_axes([0,-2.4,0.5,0.5])
axWomen.pie(mrsPresident['Women'], labels=labels, colors=colors, autopct='%.1f%%')
axWomen.set_title('Women')

plt.show()




```

Now it is easier to compare the data from yes and no more easily. All of the data is now showing aswell, with the addition of the netural data. The colors help the viewer to see more clearly the seperation of each field.

I found this image to be very overwhelming to look at. It took me awhile to figure out how the image worked and how the data was represented. The type of pie chart that was used was confusing as it was meant to show the range of the salaries. It does not make sense to show the salaries in this format as they are not completing a whole part of a pie. The boundary detection although colored was hard to make out as many salaries overlapped making it hard to see some job types. The overlap of the slices made it necessary to change the size of each slice in order for all areas to be visible. This size difference does not represent the data and is just there to make everything visible. This size difference may confuse the audience in thinking that the radius size of the slices has meaning. 


I feel as though a box plot would better represent this data as it would be able to show the ranges of the salaries without the need for overlap. It will also allow for a more consistent size and shape of the image. 

```python
salaries = {'Senior Managers': [143000,192000], 'Group Porduct Manager': [157000,172000], 'Marketing Director': [165000,179000], 'Staff User Experience': [167000,197000], 'Senior Partner Technology': [180000,195000], 'Human Resources Director': [183000,199000], 'Engineering Director': [184000,198000], 'Directors': [172000,231000], 'Product Management Director': [203000,218000], 'Lead Software Engineer Contractor': [221000,239000]}
df = pd.DataFrame(data=salaries.values(), index=salaries.keys())

df.T.boxplot(vert = 0,color='blue')
plt.title("Google Top 10 Salaries")
plt.show()
```

Now that each job type is more spread out it is much easier to consume and compare the salary ranges to one another. The boxplot as a whole is easier to read compared to the overlapping pie chart. It is also easier for the viewer to understand. Using a more conventional graph will make it easier for viewers to grasp the data alot faster.  

# Image 4

This image is extremely unclear to me. It took me awhile to figure out what it was trying to convey as the circles didn’t seem to have much meaning. The colors between clerical and non-essential functions, and time spent on professional task was unclear as the hues are very close together both on the paid and unpaid sides. The boundary detection is also poor. It impossible to tell where paid dots stop and unpaid dots begin without counting them.  I think a stacked bar chart would. do a better job of showing the time spent on certain tasks and allows for paid and unpaid to be shown side by side in a more clear and obvious state. 

```python
data = {
    "clerical and non-essential functions": np.array([25,31]),
    "professional tasks": np.array([42,30]),
    "other": np.array([(100-(25+42)), (100 - (31+30))])
}
#for this image I am assuming that paid and unpaid make up 100 for each so ive added a other coloum for the remainder.
myList = ('Paid', 'Unpaid')
fig, ax = plt.subplots()
bottom = np.zeros(2)

for boolean, myData in data.items():
    p = ax.bar(myList, myData, 0.65, label=boolean, bottom=bottom)
    ax.bar_label(p, label_type='center')
    bottom+= myData

ax.set_ylabel("% of paid/unpaid")

ax.legend(loc='upper center', bbox_to_anchor=(0.5,-0.1))

plt.show()


```

Now the data is much more clear than before. The two categories, Paid and Unpaid, are seperated now making it easy to see a boundary between the two. The addition of other also helps show a more complete view of the data. The more distinct difference in color also allows for views to spot the boundy line between time spent doing each task.
